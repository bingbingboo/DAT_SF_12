```python
#Top 10 carrier_delay by destination airports
flightdata.groupby('DEST').CARRIER_DELAY.describe().unstack().sort('count',ascending=False).head(10)

```

```python
#Top 10 carrier_delay by destination airports
flightdata.groupby('ORIGIN').CARRIER_DELAY.describe().unstack().sort('count',ascending=False).head(10)
```

```python
#Which airline has more delays?
flightdata.groupby('UNIQUE_CARRIER').CARRIER_DELAY.describe().unstack().sort('count',ascending=False).head(10)
```

```python
#If I want to fly out with Southwest, which airport will be better: SFO, OAK or SJC? The answer is OAK

flightdata1=flightdata.copy()
df=flightdata1[flightdata1['ORIGIN'].isin(['SFO','OAK','SJC'])]
df2=df[df['UNIQUE_CARRIER'].isin(['WN'])]
df2.groupby('ORIGIN').CARRIER_DELAY.describe().unstack().sort('count',ascending=False).head(10)
```
